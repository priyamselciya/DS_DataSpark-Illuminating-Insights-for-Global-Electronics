```python
#import dataset
import pandas as pd
import os
product_df = pd.read_csv(r'D:\DS_DataSpark Illuminating Insights for Global Electronics\raw_datasets\Products.csv')
```

```python
product_df
```

standardize strings

```python
product_df['Product Name'] = product_df['Product Name'].str.strip()
product_df['Brand'] = product_df['Brand'].str.strip()
product_df['Color'] = product_df['Color'].str.strip()
product_df['Subcategory'] = product_df['Subcategory'].str.strip()
product_df['Category'] = product_df['Category'].str.strip()
```

Check for null or missing values

```python
product_df.isna().sum()
```

check for duplicate rows

```python
product_df.duplicated().sum()
```

check for datatypes

```python
product_df.dtypes
```

Extract int values from price column and reassign it with float as dtype

```python
# First convert to float, then format
product_df['Unit Cost USD'] = product_df['Unit Cost USD'].str.extract(r'(\d+\.\d+|\d+)')[0].astype(float).round(2)
product_df['Unit Price USD'] = product_df['Unit Price USD'].str.extract(r'(\d+\.\d+|\d+)')[0].astype(float).round(2)
```

Rename columns

```python
product_df.columns
```

```python
product_df = product_df.rename(columns={
    'ProductKey': 'product_key',
    'Product Name': 'product_name',
    'Brand': 'brand',
    'Color': 'color',
    'Unit Cost USD': 'unit_cost_usd',
    'Unit Price USD': 'unit_price_usd',
    'SubcategoryKey': 'sub_category_key',
    'Subcategory': 'sub_category',
    'CategoryKey': 'category_key',
    'Category': 'category'
    })
```

Export cleaned Dataset

```python
# Define the directory and file path
directory = 'cleaned_dataset'
file_path = os.path.join(directory, 'products.csv')

# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Save the DataFrame to the specified file path
product_df.to_csv(file_path, index=False)
print("File saved successfully.")
```
